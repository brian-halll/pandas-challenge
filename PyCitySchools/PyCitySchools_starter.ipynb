# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

```python
# Calculate the Totals (Schools and Students)
school_count = len(school_data_complete["school_name"].unique())
student_count = school_data_complete["Student ID"].count()

# Calculate the Total Budget
total_budget = school_data["budget"].sum()
```

```python
# Calculate the Average Scores
average_math_score = school_data_complete["math_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()
```

```python
# Calculate the Percentage Pass Rates
percentagePassMath = (school_data_complete[school_data_complete["math_score"] >= 70].count()["student_name"] / \
                      student_count) * 100


percentagePassRead = (school_data_complete[school_data_complete["reading_score"] >= 70].count()["student_name"] / \
                        student_count) * 100

overall_pass_count = school_data_complete[(school_data_complete["reading_score"] >= 70) & \
                                          (school_data_complete["math_score"] >= 70)].count()["student_name"]

overall_pass_percent = (overall_pass_count / student_count) * 100

# Minor Data Cleanup
summary = pd.DataFrame(
    {
        "Total Schools" : [school_count] ,
        "Total Students" : [student_count] ,
        "Total Budget" : [total_budget] ,
        "Average Math Score" : [average_math_score] ,
        "Average Reading Score" : [average_reading_score] ,
        "% Passing Math" : [percentagePassMath] ,
        "% Passing Reading" : [percentagePassRead] ,
        "% Overall Passing" : [overall_pass_percent] ,
    }
)

# Formatting
summary["Total Students"] = summary["Total Students"].map("{:,}".format)
summary["Total Budget"] = summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
summary
#school_data_complete.head()
```

## School Summary

```python
# Determine the School Type
# Calculate the total student count

school_types = school_data_complete.set_index(['school_name'])["type"]

per_school_students = school_data_complete["school_name"].value_counts()

# Calculate the total school budget and per capita spending
# per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]

per_school_capita = per_school_budget / per_school_students

# Calculate the average test scores
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_read = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
# Get the students who passed math and passed reading by creating separate filtered DataFrames.
pass_math_df = school_data_complete[school_data_complete["math_score"] >= 70]
pass_read_df = school_data_complete[school_data_complete["reading_score"] >= 70]


# Get the the students who passed both reading and math in a separate DataFrame.
pass_both_df = school_data_complete[(school_data_complete["reading_score"] >= 70) & \
                                    (school_data_complete["math_score"] >= 70)]
```

```python
#  Calculate the Percentage Pass Rates
pass_math_per = pass_math_df.groupby(["school_name"]).count()["student_name"] / per_school_students * 100
pass_reading_per = pass_read_df.groupby(["school_name"]).count()["student_name"] / per_school_students * 100

pass_both_per = pass_both_df.groupby(["school_name"]).count()["student_name"] / per_school_students * 100
```

```python
# Convert to DataFrame
school_summary_df = pd.DataFrame(
    {
        "School Type":school_types ,
        "Total Students": per_school_students ,
        "Total School Budget":per_school_budget ,
        "Per Student Budget": per_school_capita,
        "Average Math Score": pass_math_df ,
        "Average Reading Score": pass_read_df ,
        "% Passing Math": pass_math_per ,
        "% Passing Reading":pass_reading_per ,
        "% Overall Passing":pass_both_per 
    }
)


# Minor data wrangling
school_summary_df = [["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Math Score", "Average Reading Score", 
                                         "% Passing Math", "% Passing Reading", 
                                         "% Overall Passing"]]
school_summary_df
```
